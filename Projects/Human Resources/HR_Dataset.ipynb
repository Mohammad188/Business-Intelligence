## Preprocessing

```python
# Define categories
categories = {
    'Engineering': ['Software Engineer', 'Software Engineering Manager'],
    'IT': ['IT Support', 'Sr. DBA', 'Database Administrator', 'Enterprise Architect',
           'BI Developer', 'Senior BI Developer', 'Principal Data Architect',
           'Data Architect', 'IT Manager - Support', 'Network Engineer',
           'IT Director', 'IT Manager - Infra', 'IT Manager - DB', 'CIO'],
    'Finance': ['Sr. Accountant', 'Accountant I'],
    'Sales': ['Area Sales Manager', 'Sales Manager', 'Director of Sales'],
    'Administration': ['Administrative Assistant', 'Shared Services Manager'],
    'Prodcution': ['Production Technician I', 'Production Technician II', 'Production Manager'],
    'Leadership': ['President & CEO', 'Director of Operations', 'BI Director']
}

# Function to categorize jobs
def categorize_job(title):
    for category, titles in categories.items():
        if title in titles:
            return category
    return 'Other'

# Create a new column
df['Category'] = df['Position'].apply(categorize_job)
```

```python
df[['Position', 'Category']]
```

```python
df = df.rename(columns={'Category': 'Job Category'})
```
